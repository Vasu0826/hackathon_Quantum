```python
from pywinauto import Desktop
import time

def find_all_sub_children(window_title):
    # Get the handle of the window
    desktop = Desktop(backend="uia")
    window = desktop.window(title=window_title)
    
    # Wait for the window to be ready (adjust delay as needed)
    time.sleep(1)
    
    # Initialize a list to store all sub-children
    all_sub_children = []
    
    # Recursive function to traverse the control tree and find all sub-children
    def traverse_children(parent):
        children = parent.children()
        for child in children:
            all_sub_children.append(child)
            traverse_children(child)
    
    # Start traversal from the top-level window
    traverse_children(window)
    
    return all_sub_children

# Example usage
if __name__ == "__main__":
    # Title of the window
    window_title = "YouTube Downloader HD - High Speed Edition"
    
    # Find all sub-children within the window
    all_sub_children = find_all_sub_children(window_title)
    
    # Print information about each sub-child
    for sub_child in all_sub_children:
        print("Control ID:", sub_child.control_id())
        print("Class Name:", sub_child.class_name())
        print("Text:", sub_child.window_text())
        print("Rectangle:", sub_child.rectangle())
        print()
```

```python
from pywinauto import Desktop

def send_keys_by_control_id(window_title, control_id1, control_id2, keys1, keys2):
    # Get the handle of the window
    desktop = Desktop(backend="uia")
    window = desktop.window(title=window_title)
    
    # Find the control by its control ID
    control = window.child_window(control_id=control_id1)
    
    # Send keys to the control
    control.type_keys(keys1)

    
    # Find the control by its control ID
    control = window.child_window(control_id=control_id2)
    
    # Send keys to the control
    control.type_keys(keys2)

# Example usage
if __name__ == "__main__":
    # Title of the window
    window_title = "YouTube Downloader HD - High Speed Edition"
    
    # Control ID of the target control
    control_id1 = 263804
    control_id2 = 460260
    
    # Keys to send
    keys1 = "example_keys1"
    keys2 = "example_keys2"
    
    # Send keys to the window based on control ID
    send_keys_by_control_id(window_title, control_id1, control_id2, keys1, keys2)
```
